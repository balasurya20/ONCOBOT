```python
import matplotlib.pyplot as plt
import numpy as np
# Data for audio classification execution times
audio_environments = ["With Intel OneAPI", "Without Intel OneAPI"]
audio_classification_times = [3.7, 4.9]  # Replace these with your desired execution times

# Create a stylish bar chart
plt.figure(figsize=(8, 6))  # Set figure size

# Create the bars with custom color gradient
colors = plt.cm.get_cmap('YlGnBu')(np.linspace(0.3, 0.7, len(audio_environments)))  # Color gradient
plt.bar(audio_environments, audio_classification_times, color=colors)

# Customize labels and title with font styles
plt.xlabel("Execution Environment", fontweight='bold', fontsize=12)
plt.ylabel("Time (seconds)", fontweight='bold', fontsize=12)
plt.title("Oncology based OCR Model Execution Comparison\n(With vs. Without Intel OneAPI)", fontweight='bold', fontsize=14)

# Add grid with a specific linestyle
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Display data values on top of bars
for i, v in enumerate(audio_classification_times):
    plt.text(i, v + 0.1, f"{v:.2f}s", ha='center', va='bottom', fontsize=11, fontweight='bold')  # Adjust y-offset for better positioning

# Remove unnecessary spines
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()

# Import numpy for color gradient (assuming it's not already installed)
import numpy as np
```
